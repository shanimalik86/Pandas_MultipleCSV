```python
from pathlib import Path
import pandas as pd
import numpy as np

path = r'/home/zeeshanfazal/fate-bioinformatics/xCL code/Targeted_Data'  # or unix / linux / mac path

# Get the files from the path provided in the OP
files = Path(path).glob('*.csv')  # .rglob to get subdirectories

dfs = list()
for f in files:
    data = pd.read_csv(f)
    data=data[data.Log2FoldChange>=2]
    # .stem is method for pathlib objects to get the filename w/o the extension
    data['file'] = f.stem
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)
#df=df[df.Log2FoldChange>=2]
df.to_csv('All_TargetData_LFC2_test.csv')
```
